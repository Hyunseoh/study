```python
titanic_train=pd.read_csv('train.csv')
titanic_test=pd.read_csv('test.csv')
titanic_train.info()
titanic_train
```

```python
#*타이타닉 데이터셋(캐글에서 다운받아 진행)*
#1) sort_values 메서드를 사용하여 타이타닉호 승객에 대해 
#성별(sex) 인원수, 나이별(age) 인원수, 선실별(pclass) 인원수, 사망/생존(survived) 인원수를 구하라.
titanic_train.sort_values(by='Sex')
titanic_train.sort_values(by='Age')
titanic_train.sort_values(by='Pclass')
titanic_train.sort_values(by='Survived')

#2)타이타닉호 승객의 평균 나이를 구하라.
titanic_train['Age'].mean()

#3)타이타닉호 승객중 여성 승객의 평균 나이를 구하라.
titanic_train[titanic_train.Sex=='female']['Age'].mean()
#4)타이타닉호 승객중 1등실 선실의 여성 승객의 평균 나이를 구하라.
titanic_train[(titanic_train['Sex']=='female')& (titanic_train.Pclass==1)]['Age'].mean()
#5) 타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 category1 열을 만들어라.
#category1 카테고리는 다음과 같이 정의된다.
titanic_train['category1']=np.nan
#- 20살이 넘으면 성별을 그대로 사용한다.
#- 20살 미만이면 성별에 관계없이 “child”라고 한다.
titanic_train.category1=titanic_train.apply(lambda x: x.Sex if x.Age>20 else 'child', axis=1 )
```

```python
#6)타이타닉호의 승객 중 나이를 명시하지 않은 고객은 나이를 명시한 
#고객의 평균 나이 값이 되도록 titanic 데이터프레임을 고쳐라.
titanic_train.Age[titanic_train['Age'].isnull()]=round(titanic_train['Age'].mean(),0)
#A value is trying to be set on a copy of a slice from a DataFrame
#와 같은 경고메시지 뜨지만 일단 평균값으로 NaN이 대체됨.. ---> .copy() 로 복사본 만들어서 함수 적용하면
#에러 메시지 뜨지 않는다고 한다. 
titanic_train.Age.dtype
titanic_train.Age=titanic_train.Age.astype(np.int64)

#7)타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 category2 열을 만들어라. 
#category2 카테고리는 다음과 같이 정의된다.
#- 성별을 나타내는 문자열 male 또는 female로 시작한다.
#- 성별을 나타내는 문자열 뒤에 나이를 나타내는 문자열이 온다.
#- 예를 들어 27살 남성은 male27 값이 된다.

titanic_train['category2']=titanic_train.Sex+titanic_train.Age.astype(str)

#8)타이타닉호 승객을 ‘미성년자’, ‘청년’, ‘중년’, ‘장년’, ‘노년’ 나이 그룹으로 나눈다.
#bins = [1, 20, 30, 50, 70, 100]
#labels = ["미성년자", "청년", "중년", "장년", "노년"]
#그리고 각 나이 그룹의 승객 비율을 구한다. 비율의 전체 합은 1이 되어야 한다.

titanic_train['Ageg']=pd.cut(titanic_train.Age,
   bins = [1, 20, 30, 50, 70, 100],
labels = ["미성년자", "청년", "중년", "장년", "노년"])

titanic_train['Ageg'].value_counts()/len(titanic_train['Ageg'])
(titanic_train['Ageg'].value_counts()/len(titanic_train['Ageg'])).sum()
#0.984287317620651 #1이 안됨

sum(titanic_train.Ageg.isnull())
#결측값 14개
np.where(titanic_train.Ageg.isnull())#결측값 위치 파악
titanic_train['Age'][[ 78, 164, 172, 183, 305, 381, 386, 469, 644, 755, 788, 803, 827,
        831]]
#0살과 1살이 포함되지 않음

titanic_train['Ageg']=pd.cut(titanic_train.Age,
   bins = [-1, 20, 30, 50, 70, 100],
labels = ["미성년자", "청년", "중년", "장년", "노년"])
#bins조정
titanic_train['Ageg'].value_counts()/len(titanic_train['Ageg'])
(titanic_train['Ageg'].value_counts()/len(titanic_train['Ageg'])).sum()
```

```python

#9)타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 category3 열을 만들어라. 
#category3 카테고리는 다음과 같이 정의된다.
#- 20살 미만이면 성별에 관계없이 “미성년자”라고 한다.
#- 20살 이상이면 나이에 따라 “청년”, “중년”, “장년”, “노년”을 구분하고 
#그 뒤에 성별을 나타내는 “남성”, “여성”을 붙인다.
titanic_train['category3']=titanic_train.apply(lambda x: '미성년자' if x.Age < 20 else x.Ageg+x.Sex, axis=1)

#10) 생존율 비교
#- Pclass/Sex/SibSp/Parch/Embarked 에 따른 생존율

for i in ['Pclass','Sex','SibSp','Parch','Embarked']:
    by_survive=titanic_train.groupby([i]).Survived.value_counts()
    print(by_survive/by_survive.sum())
```

```python
#3) 유클리디안 거리 알고리즘을 이용하여 생존 여부 판단
#- test 데이터셋의 Pclass, Sex, Age, SibSp, Parch, Fare 열 조합
#--->KNN 알고리즘 이용하기로 함 


train=titanic_train[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
test=titanic_test[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]

labels=titanic_train['Survived']
```

```python
#(트레이닝)결측값과 데이터타입 확인
for i in list(train.columns):
    print(i, train[i].isnull().sum(),train[i].dtype )
#결측값: Age 177    
    
#(테스트)결측값과 데이터타입 확인
for i in list(test.columns):
    print(i, test[i].isnull().sum(),test[i].dtype )
#결측값: Age 86 Fare 1
#테스트데이터의 Fare열은 결측치 하나뿐이므로 그냥 중위수로 채워줌
test.Fare.fillna(titanic_test.Fare.median(), inplace=True)
```

```python
sum(test['Fare'].isnull())
```

```python
#거리형 연산을 위해 Sex열 변환

onehot_sex =pd.get_dummies(train['Sex'])
train= pd.concat([train,onehot_sex], axis=1)
#female과 male열을 만들고 값을 0,1로 채움
del train['Sex']
onehot_sex =pd.get_dummies(test['Sex'])
test= pd.concat([test,onehot_sex], axis=1)
del test['Sex']
```

```python
#나이 NA채우기 여러 방법으로 해보기 위해 Age_na 인덱스 저장해놓기로 함

train_AgeNA_idx=list(train[train.Age.isnull()].index)
test_AgeNA_idx=list(test[test.Age.isnull()].index)
```

```python

#1. 전체 승객 나이 평균으로 대체
train.Age.fillna(round(train.Age.mean(),0), inplace=True)
test.Age.fillna(round(test.Age.mean(),0), inplace=True)
#경고창, 변경은 됨
sum(train.Age.isna())
```

```python
#변수들 표준화

#np.mean(list(train.Age))
#np.std(list(train.Age))
def stdzation(x):
    std_lst=[]
    _mean_=np.mean(list(x))
    _std_=np.std(list(x))
    for i in x:
        std_lst.append((i-_mean_)/_std_)
    return std_lst
```

```python
imsi={}
for col in train.columns:
    imsi[col]=stdzation(train[col])
imsi2={}
for col in test.columns:
    imsi2[col]=stdzation(test[col])
```

```python
train_std=pd.DataFrame(imsi)
test_std=pd.DataFrame(imsi2)
```

```python
from sklearn.preprocessing import StandardScaler
train_std=StandardScaler().fit_transform(train)
test_std=StandardScaler().fit_transform(test)
#하면 된다함... 
```

```python
from sklearn.neighbors import KNeighborsClassifier as knn_c
from sklearn.neighbors import KNeighborsRegressor as knn_r
```

```python
knn= knn_c(n_neighbors=10)
```

```python
#모형 학습
knn.fit(train_std,labels)
```

```python
test_survive=knn.predict(test_std)
```
